# Qs 1 .What was the best month for sales ? How much was earned that month?

```python
df_filtered["Quantity Ordered"]=df_filtered["Quantity Ordered"].astype("float")
df_filtered["Price Each"]=df_filtered["Price Each"].astype("float")       
```

```python
df_filtered.info()
```

```python
df_filtered["Price Each"].nunique
```

```python
df_filtered["sales"]=df_filtered\
["Quantity Ordered"]*df_filtered["Price Each"]
```

```python
df_filtered.head()
```

```python
df_filtered["month"].unique()
```

```python
df_filtered.info()
```

```python
df=df_filtered.groupby(['month'])\
['sales'].sum().sort_values(ascending=False)
df=df.to_frame()
```

```python
df.info
```

```python
df["sales"]
```

```python
df.columns
```

```python
month=["dec","oct","apr","nov",
       "may","mar","july","june"
       ,"aug",'feb',"sep","jan"] 
df["month"]=month
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

sns.barplot(x = "sales",
            y = "month",
            data = df)
plt.title("Month wise Sale")
plt.show()
```

Best Month for sales  was DECEMBER
